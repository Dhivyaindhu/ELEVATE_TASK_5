```python
import numpy as np
import pandas as pd
data=pd.read_excel('/content/Product-Sales-Region.xlsx')
data.describe()
data.info()
```

```python
data.isnull().sum()
```

```python
data['Promotion']=data["Promotion"].fillna('Unknown')
data.head(10)
```

Univariate Analysis for the Categorical Values

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='RegionManager',data=data)
plt.title("Region Manager")
plt.show()
```

```python
import seaborn as sns
sns.countplot(x='PaymentMethod',data=data)
plt.title("Payment Method Over the data")
plt.show()
```

Univariate Analysis for the Numerical Values

```python
data['UnitPrice'].hist(bins=20,edgecolor='blue')
plt.title("Unit Price Distribution")
plt.xlabel("Unit Price")
plt.ylabel("Frequency")
plt.show()
```

Bivariate Analysis

```python
crosstab=pd.crosstab(data['Product'],data['Region'])
sns.heatmap(crosstab,annot=True)
plt.title("Product vs Region")
plt.show()
```

```python
sns.violinplot(x='CustomerName',y='OrderID',data=data)
plt.title("Customer vs Order")
plt.show()
```

Multivariate Analysis

```python
import seaborn as sns
sns.boxplot(x='Product',y='TotalPrice',hue='StoreLocation',data=data)
plt.title("Product vs Total Price")
plt.show()
```

```python
crosstab=pd.crosstab(index=data['StoreLocation'],columns=[data['CustomerName'],data['Product']])
sns.heatmap(crosstab,annot=True,cmap='YlGnBu')
plt.title("Orders On Store Location")
plt.show()
```

```python
sns.scatterplot(x='UnitPrice',y='TotalPrice',hue='Product',data=data)
plt.title("Unit Price vs Total Price")
plt.show()
```
